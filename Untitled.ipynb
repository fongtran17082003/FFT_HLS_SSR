```python
import numpy as np
import math
from pynq import Overlay, allocate
ol = Overlay("fft_ssr.bit")
```

```python
my_ip = ol.fft_4k_ssr2_i16_c16_0  # đổi tên nếu IP khác
```

```python
def generate_test_data(fft_size, iter, dtype_bits=16):

    qpt = dtype_bits - 2
    data = np.zeros(fft_size * iter, dtype=np.complex64)

    for i in range(iter):

        step = fft_size // iter if fft_size >= iter else 1
        phase_factor = 2.0 * math.pi * step / fft_size

        for k in range(fft_size):

            re = math.cos(phase_factor * k)
            im = math.sin(phase_factor * k)

            scale = 2 ** qpt

            data[i*fft_size+k] = complex(
                int(re * scale),
                int(im * scale)
            )

    return data
```

```python
def pack_input(din, FFT_SIZE, SSR, input_width, iter, dtype_bits=16):

    qpt_in = dtype_bits - 2
    qpt_out = input_width - 2

    conv_factor = 2 ** (qpt_out - qpt_in)

    total = FFT_SIZE // SSR * iter

    packed = np.zeros(total, dtype=np.int64)

    for i in range(total):

        temp = 0

        for k in range(SSR):

            idx = SSR*i + k

            real = int(din[idx].real * conv_factor)
            imag = int(din[idx].imag * conv_factor)

            temp |= (real & ((1<<input_width)-1)) << (2*k*input_width)
            temp |= (imag & ((1<<input_width)-1)) << ((2*k+1)*input_width)

        packed[i] = temp

    return packed

```

```python
FFT_SIZE = 4096
SSR = 2
INPUT_WIDTH = 16
ITER = 1

data = generate_test_data(FFT_SIZE, ITER)

packed = pack_input(data, FFT_SIZE, SSR, INPUT_WIDTH, ITER, dtype_bits=16)

# packed = pack_debug(data, SSR, INPUT_WIDTH)

print(packed)
```

```python
num_words = FFT_SIZE // SSR * ITER

input_buffer = allocate(shape=(num_words,), dtype=np.int64)
output_buffer = allocate(shape=(num_words,), dtype=np.int64)
output_buffer = allocate(shape=(num_words,), dtype=np.int64)

print("Input buffer address :", hex(input_buffer.device_address))
print("Output buffer address:", hex(output_buffer.device_address))
```
